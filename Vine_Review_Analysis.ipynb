```python
import pandas as pd
import numpy as np
```

```python
data = ('Vine.csv')
df = pd.read_csv(data)
df.head()
```

```python
# create a new DataFrame to retrieve all the rows where the total_votes count is equal to or greater than 20 
total_votes_df = df.loc[df["total_votes"] >= 20]
total_votes_df
```

```python
# Retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
helpful_votes_df = total_votes_df[(total_votes_df['helpful_votes']/total_votes_df['total_votes']) >= 0.5]
helpful_votes_df
```

```python
# retrieves all the rows where a review was written as part of the Vine program (paid)
paid_df = helpful_votes_df.loc[helpful_votes_df["vine"] == 'Y']
paid_df
```

```python
#  retrieve all the rows where the review was not part of the Vine program (unpaid)
unpaid_df = helpful_votes_df.loc[helpful_votes_df["vine"] == 'N']
unpaid_df
```

```python
# the total number of reviews for paid
total_paid_df = paid_df.['review_id'].count()
total_paid_df
```

```python
# the total number of reviews for unpaid
total_unpaid_df = unpaid_df['review_id'].count()
total_unpaid_df
```

```python
# the number of paid 5-star reviews 
paid_5_star = len(paid_df[(paid_df['star_rating'] == 5)])
paid_5_star
```

```python
# the number of unpaid 5-star reviews
unpaid_5_star = len(unpaid_df[(unpaid_df['star_rating'] == 5)])
unpaid_5_star
```

```python
#  the percentage of paid 5-star reviews
percent_paid = paid_5_star / total_paid_df * 100
percent_paid
```

```python
#  the percentage of unpaid 5-star reviews
percent_unpaid = unpaid_5_star / total_unpaid_df * 100
percent_unpaid
```
